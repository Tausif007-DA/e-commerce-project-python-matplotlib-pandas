```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(r"C:\Users\raza0\Downloads\ecommerce_visualization_dataset_1000_rows (1).xlsx")

df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Month']=df['Date'].dt.month
df['Year']=df['Date'].dt.year
```

```python
Total_Revenue=df['Revenue'].sum()
Total_Profit=df['Profit'].sum()
Avg_Rating=df['Customer_Rating'].mean()
```

```python
print("Total Revenue",Total_Revenue)
print("Total _profit",Total_Profit)
print("Average customer rating",Avg_Rating)
```

# revenue trend

```python
monthly_revenue =df.groupby('Month')['Revenue'].sum()

plt.figure(figsize=(8,4))

plt.plot(monthly_revenue)

plt.title("MONTHLY REVENUE TREND")

plt.xlabel("Month")
plt.ylabel("Revenue")

plt.xticks(range(1,13))
plt.grid()
plt.show()
```

# category wise trend

```python
category_revenue = df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,4))

plt.bar(category_revenue.index,category_revenue.values)

plt.title("Category wise revenue")
plt.show()
```

# Region-wise Revenue

```python
regional_revenue =df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,4))

plt.bar(regional_revenue.index,regional_revenue.values)

plt.title("Region wise revenue")
plt.xlabel("region")
plt.ylabel("revenue")
plt.show()
```

# Marketing vs Revenue 

```python
plt.figure(figsize=(8,4))
plt.scatter(df["Marketing_Spend"],df["Revenue"],alpha =0.3)

plt.title("Marketing vs Revenue")
plt.xlabel("marketing")
plt.ylabel("Revenue")
plt.show()
```

# Profit Distribution

```python
plt.figure(figsize=(8,4))
plt.boxplot(df["Profit"])
plt.title("Profit Distribution")
plt.show()
```

1. The business shows fluctuating revenue trends across months, indicating no stable growth pattern.

2. Certain categories contribute significantly more revenue, suggesting a focus area for business expansion.

3. Regional analysis highlights uneven performance, with some regions underperforming.

4. Scatter analysis shows weak correlation between marketing spend and revenue, indicating inefficiencies in marketing strategy.

5. Profit distribution is highly variable with multiple outliers, suggesting inconsistent profitability.
